```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import webdriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.chrome()
driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")
action = ActionChains(driver)
route_collected=[]
for i in range(1,6):
    b=driver.find_element(By.CSS_SELECTOR,f".DC_117_paginationTable .DC_117_pageTabs:nth-child({i})")
    action.move_to_element(b).click().perform()
    c=driver.find_elements(By.CLASS_NAME,"route")
    route_collected.extend([i.text for i in c])
page1=driver.find_element(By.CSS_SELECTOR,f".DC_117_paginationTable .DC_117_pageTabs:nth-child(i)")    
action.move_to_element(page1).click().perform()
d = driver.find_element(By.LINK_TEXT,"vijayawada to hyderabad")
action.move_to_element(d).click().perform()
time.sleep(5)
wait=webdriverWait(driver,100)
e=wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR,"#result-section > div:nth-child(1).button")))
action.move_to_element(e).click().perform()
element = driver.find_element(By.CSS_SELECTOR,"#result-section > div:nth-child(2).button")
driver.execute_script("arguments[0].scrollIntoView();",element)
action.move_to_element(element).click().perform()
element = webdriverWait(driver,30).until(EC.presence_of_all_elements_located(By.XPATH,'//div[@class="clearfix bus-item"]'))
body = driver.find_element(By.TAG_NAME,'body')
while True:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if driver.execute_script("return window.innerheight + window.scrollY") >= new_height:
        break
buses = driver.find_elements(By.XPATH,'//div[@class="clearfix bus-item-details"]')
bus_details=[]
for bus in buses:
    name = bus.find_element(By.CSS_SELECTOR,".travels").text    
    type =bus.find_element(By.CSS_SELECTOR,".bus_type").text
    departure_time = bus.find_element(By.CSS_SELECTOR,".dp_time").text
    arrival_time = bus.find_element(By.CSS_SELECTOR,".bp_time").text
    duration = bus.find_element(By.CSS_SELECTOR,".dur").text
    price =bus.find_element(By.CSS_SELECTOR,".seat-fare").text
    seats_available = bus.find_element(By.CSS_SELECTOR,"seat-left").text
    try:
        rating = bus.find_element(By.CSS_SELECTOR,".rating-sec .rating").text
    except:
        rating =" No rating"

bus_details.append({
    "name":name,
    "type":type,
    "departure_time":departure_time,
    "arrival_time" :arrival_time,
    "duration" : duration,
    "price" : price,
    "seats_available" : seats_available,
    "rating" : rating
})        




driver=webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")
action=ActionChains(driver)
route_collected =[]
c = driver.find_element(By.CLASS_NAME,"route")
route_collected.extend([i.text for i in c])
for i in route_collected:
    d = driver.find_element(By.LINK_TEXT,i)
    action.move_to_element(d).click().perform()
    bus_details()
time.sleep(10)
     
```

```python

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Set up the Selenium WebDriver 
driver = webdriver.Chrome()

# Open the Redbus website
driver.get("https://www.redbus.in/")

# Add a wait time for the page to load
time.sleep(5)

# Fill in the source and destination fields, then search
source = driver.find_element(By.ID, 'src')
source.send_keys('Bangalore')

destination = driver.find_element(By.ID, 'dest')
destination.send_keys('Chennai')

search_button = driver.find_element(By.ID, 'search_btn')
search_button.click()

# Wait for the results to load
time.sleep(10)

# Scrape the results (this shows the structure of the results page; adjust selectors as needed)
bus_names = driver.find_elements(By.CLASS_NAME, 'travels')
departure_times = driver.find_elements(By.CLASS_NAME, 'dp-time')
arrival_times = driver.find_elements(By.CLASS_NAME, 'bp-time')
durations = driver.find_elements(By.CLASS_NAME, 'dur')
prices = driver.find_elements(By.CLASS_NAME, 'fare')

# Extract data
data = []
for i in range(len(bus_names)):
    data.append({
        'Bus Name': bus_names[i].text,
        'Departure Time': departure_times[i].text,
        'Arrival Time': arrival_times[i].text,
        'Duration': durations[i].text,
        'Price': prices[i].text
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Close the driver
driver.quit()
```
